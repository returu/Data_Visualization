```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# 设置中文字体
plt.rcParams['font.sans-serif']=['SimHei']
#显示负号
matplotlib.rcParams['axes.unicode_minus']=False
```

# 1、函数语法：

```python
from matplotlib.patches import Annulus

fig, ax = plt.subplots(figsize=(6, 6))

# 创建圆环
annulus = Annulus(
    xy=(0.5, 0.5),         # 椭圆环中心点坐标
    r=(0.3, 0.4),          # 外轮廓的半轴长度(短半轴, 长半轴)
    width=0.2,             # 环的厚度，从外轮廓向内收缩
    angle=45,              # 椭圆环的旋转角度（逆时针方向）
    facecolor='red',       # 环的填充颜色
    edgecolor='black',     # 环的边框颜色
    linewidth=2,           # 边框线的宽度
    linestyle='--',        # 边框线的样式
    hatch='/',             # 填充图案
    alpha=0.4              # 整体透明度
                 )

# 将圆环添加到坐标轴
ax.add_patch(annulus)

ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')  # 确保比例相等
ax.set_title('基本椭圆环示例')
plt.show()
```

# 2、使用示例：

### 示例 1：创建靶心图

```python
from matplotlib.patches import Annulus

fig, ax = plt.subplots(figsize=(6, 6))

# 定义环的颜色
colors = ['red', 'white', 'blue', 'white', 'red']

# 创建多个同心圆环
for i in range(5):
    annulus = Annulus(xy=(0.5, 0.5), 
                      r=0.4 - i*0.06, 
                      width=0.06,
                      facecolor=colors[i],
                      edgecolor='black',
                      alpha=0.6
                     )
    ax.add_patch(annulus)

ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')  # 确保比例相等
ax.set_title('靶心图', fontsize=15)
plt.show()
```
