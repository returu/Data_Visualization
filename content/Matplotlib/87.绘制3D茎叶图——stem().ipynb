```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# 设置中文字体
plt.rcParams['font.sans-serif']=['SimHei']
#显示负号
matplotlib.rcParams['axes.unicode_minus']=False
```

# 1、函数语法：

```python
# 创建更复杂的数据
t = np.linspace(0, 10, 50)
x = np.cos(t)
y = np.sin(t)
z = t

fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

# 自定义样式绘制
markerline, stemlines, baseline = ax.stem(
    x, y, z,             # 顶端标记点的位置坐标
    linefmt='C1--',      # 茎：青色虚线
    markerfmt='C2^',     # 叶：绿色三角形标记
    basefmt='C4-',       # 基准线：蓝色实线
    bottom=-1,           # 基线位置
    label='3D Stem Data' # 图例标签
)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D茎叶图')

plt.legend()
plt.show()
```

# 2、使用示例：

### 示例 1：方向控制（orientation参数）

```python
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

# 生成示例数据
n = 15
x = np.random.rand(n)
y = np.random.rand(n) 
z = np.arange(n)

# 水平方向茎叶图
markerline, stemlines, baseline = ax.stem(
    x, y, z,
    orientation='x',    # 沿x方向投影
    bottom=-1           # 基线位置
)

ax.set_title('水平方向的3D茎叶图')

plt.show()
```

### 示例 2：数据裁剪（axlim_clip参数）

```python
# 生成螺旋数据
z = np.arange(0, 4*np.pi, np.pi/20)
x = z * np.cos(z)
y = z * np.sin(z)

fig = plt.figure(figsize=(16, 6))

# 第一个子图：未进行数据裁剪 (axlim_clip=False)
ax1 = fig.add_subplot(121, projection='3d')
ax1.stem(x, y, z, 
         axlim_clip=False,  # 显示完整茎线，即使超出坐标轴范围
         label='未裁剪'
        )
ax1.set_xlabel('X轴', fontsize=11)
ax1.set_ylabel('Y轴', fontsize=11)
ax1.set_zlabel('Z轴', fontsize=11)
ax1.set_xlim([-8, 8])
ax1.set_ylim([-8, 8])
ax1.set_zlim([0, 10])
ax1.set_title('未裁剪的3D茎叶图', fontsize=14)
ax1.legend()

# 第二个子图：进行数据裁剪 (axlim_clip=True)
ax2 = fig.add_subplot(122, projection='3d')
ax2.stem(x, y, z, 
         axlim_clip=True,   # 仅显示坐标轴范围内的茎线
         label='已裁剪'
        )
ax2.set_xlabel('X轴', fontsize=11)
ax2.set_ylabel('Y轴', fontsize=11)
ax2.set_zlabel('Z轴', fontsize=11)
ax2.set_xlim([-8, 8])
ax2.set_ylim([-8, 8])
ax2.set_zlim([0, 10])
ax2.set_title('裁剪后的3D茎叶图', fontsize=14)
ax2.legend()

plt.show()
```
