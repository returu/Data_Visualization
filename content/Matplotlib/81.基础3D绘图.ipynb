```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# 设置中文字体
plt.rcParams['font.sans-serif']=['SimHei']
#显示负号
matplotlib.rcParams['axes.unicode_minus']=False
```

# 1、初始化3D绘图环境

```python
# 创建画布
fig = plt.figure(figsize=(6, 6))

# 添加3D坐标轴
ax = fig.add_subplot(projection='3d')

# 设置坐标轴标签
ax.set_xlabel('X轴',color='red') 
ax.set_ylabel('Y轴',color='red')
ax.set_zlabel('Z轴', labelpad=0, rotation=90,color='red')

ax.set_zlim(0,2)
ax.grid(alpha=0.3)
ax.set_title('初始化3D绘图环境')

plt.show()
```

# 2、视角控制

```python
# 创建3D图形窗口
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')

# 设置坐标轴标签
ax.set_xlabel('X',labelpad=-3,color='red') 
ax.set_ylabel('Y',color='red')
ax.set_zlabel('Z',color='red')

# 设置视角
ax.view_init(elev=20., azim=-35, roll=0,vertical_axis='x')

ax.set_title('视角控制')

plt.show()
```

# 3、使用示例：

### 示例 1：3D折线图

```python
# 创建3D图形窗口
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')

# 创建螺旋线数据
z = np.linspace(0,1,500)
x = z * np.sin(50*z)
y = z * np.cos(50*z)

# 绘制3D折线图
ax.plot3D(x, y, z, 'red', label='螺旋线',lw=5)
# ax.plot(x, y, z, 'red', label='螺旋线',lw=5) # 可等效替代

ax.set_xlabel('X轴',color='blue') 
ax.set_ylabel('Y轴',color='blue')
ax.set_zlabel('Z轴', labelpad=0, rotation=90,color='blue')
plt.legend()
plt.title("3D折线图")
plt.show()
```

### 示例 2：3D散点图

```python
# 创建数据
z = np.linspace(0,1,500)
x = z * np.sin(50*z)
y = z * np.cos(50*z)
c = x + y*2

# 创建3D图形窗口
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')

# 绘制3D散点图
sc = ax.scatter3D(x, y, z, c=c, cmap='hsv')
# sc = ax.scatter(x, y, z, c=c, cmap='hsv')   # 可等效替代

# 设置轴标签
ax.set_xlabel('x',fontsize=14,color='b')
ax.set_ylabel('y',fontsize=14,color='b')
ax.set_zlabel('z',fontsize=14,color='b')
# 添加颜色条：与散点图的颜色映射关联，shrink=0.5使颜色条长度为原图的50%
plt.colorbar(sc ,shrink=0.5) 

plt.title("3D散点图")
plt.show()
```

### 示例 3：3D等高线图

```python
# 创建数据网格
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# 创建图形和3D坐标轴
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# 绘制三维等高线图
# X, Y, Z：网格数据和对应的高度值
# levels=20：指定等高线的层级数量为20
# cmap='viridis'：使用viridis颜色映射方案
ax.contour3D(X, Y, Z, levels=20, cmap='viridis')
# ax.contour(X, Y, Z, levels=20, cmap='viridis')  # 可等效替代

# 设置坐标轴标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.title("3D等高线图")
plt.show()
```
