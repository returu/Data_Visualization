```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# 设置中文字体
plt.rcParams['font.sans-serif']=['SimHei']
#显示负号
matplotlib.rcParams['axes.unicode_minus']=False
```

# 1、函数语法：

### text()和text3D()：

```python
# 创建数据
np.random.seed(42)
x = np.random.rand(20) * 10
y = np.random.rand(20) * 10
z = np.random.rand(20) * 10

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 绘制散点图
scatter = ax.scatter(x, y, z, c=z, cmap='viridis', s=100, alpha=0)

# ===== 不同zdir参数的演示 =====

# 默认朝向 (zdir='z' 或 None)
ax.text3D(1, 5, 5, '朝向Z轴 (默认)', 
         zdir='z', color='red', fontsize=12, fontweight='bold')

# 朝向X轴
ax.text3D(10, 5, 5, '朝向X轴', 
         zdir='x', color='blue', fontsize=12, fontweight='bold')

# 朝向Y轴  
ax.text3D(8, 2, 5, '朝向Y轴',
         zdir='y', color='green', fontsize=12, fontweight='bold')

# 自定义方向向量
ax.text3D(7, 7, 7, '自定义方向',
         zdir=(1, 1, 1), color='purple', fontsize=12, fontweight='bold')

# 添加坐标轴标签
ax.set_xlabel('X轴')
ax.set_ylabel('Y轴') 
ax.set_zlabel('Z轴')

plt.title('zdir参数', fontsize=12)
plt.show()
```

### text2D()：

```python
# 创建图形窗口，并添加3D子图
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 在坐标轴坐标系中添加文本
ax.text2D(0.5, 0.5, '在坐标轴坐标系中添加文本',transform=ax.transAxes, color='blue')

# 在数据坐标系中添加文本（不指定transform参数，使用默认数据坐标系）
ax.text2D(0.05, 0.05, '在数据坐标系中添加文本', 
         ha='center',
         color='red')

# 在图形坐标系中添加文本
ax.text2D(0.5, 0.95, '在图形坐标系中添加文本',
         transform=fig.transFigure,
         ha='center',
         fontsize=12, fontweight='bold')

plt.show()
```

# 2、使用示例：

```python
# 创建数据
np.random.seed(42)
x = np.random.rand(40) * 10
y = np.random.rand(40) * 10
z = np.random.rand(40) * 10

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 绘制散点图
scatter = ax.scatter(x, y, z, c=z, cmap='viridis', s=200)

# =====使用text3D添加3D标注=====
for i in range(len(x)):
    if i % 5 == 0:  # 每隔5个点标注一次
        ax.text3D(x[i], y[i], z[i], f'P{i}', 
                 fontsize=14, color='darkred')

# =====使用text2D添加固定标注=====
ax.text2D(0.02, 0.98, '3D数据分布可视化', transform=ax.transAxes,
         fontsize=16, fontweight='bold')
ax.text2D(0.02, 0.93, '颜色表示Z值大小', transform=ax.transAxes,
         bbox=dict(boxstyle="round,pad=0.3", facecolor="orange", alpha=0.3))

# 添加颜色条
cbar = fig.colorbar(scatter, ax=ax,shrink=0.5, aspect=5)
cbar.set_label('Z值', rotation=270, labelpad=15)

# 设置坐标轴标签
ax.set_xlabel('X轴')
ax.set_ylabel('Y轴')
ax.set_zlabel('Z轴')

plt.tight_layout()
plt.show()
```
