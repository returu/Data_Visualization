```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

# 1、函数语法：

```python
from matplotlib.font_manager import FontProperties
import os

# 配置xkcd默认英文手绘字体：Humor Sans（已安装在C:/Windows/Fonts）
humor_sans_path = "C:/Windows/Fonts/Humor-Sans.ttf"
humor_font = FontProperties(fname=humor_sans_path, size=11)

# --------------------------------------------------
# 进入xkcd上下文，应用手绘风格（使用Humor Sans字体）
# --------------------------------------------------
with plt.xkcd(scale=3, length=50, randomness=4):
    fig = plt.figure(figsize=(10, 6), dpi=100)
    # 自定义坐标轴位置
    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))  

    # 设置图表基础样式（隐藏顶部/右侧边框，清空刻度）
    ax.spines[['top', 'right']].set_visible(False)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylim([-30, 10])

    # 生成数据
    data = np.ones(100)
    data[70:] -= np.arange(30)

    # 添加核心标注
    ax.annotate(
        text='THE DAY I REALIZED\nI COULD COOK BACON\nWHENEVER I WANTED',
        xy=(70, 1),       
        xytext=(15, -10),      # 文本位置
        fontproperties=humor_font, 
        fontsize=12,
        arrowprops=dict(
            arrowstyle='->',   # 实心箭头，匹配xkcd手绘感
            color='#333333',   # 箭头颜色（深灰，不刺眼）
            linewidth=1.5,
            connectionstyle='arc3,rad=0.1'  # 轻微弧度，更自然
        ),
        ha='center'  # 文本水平居中
    )

    # 绘制健康曲线（xkcd风格自动生效，线条抖动）
    ax.plot(data, color='#2E86AB', linewidth=2.5)

    # 设置坐标轴标签
    ax.set_xlabel(
        'TIME',
        fontproperties=humor_font,
        fontsize=12,
        labelpad=8
    )
    ax.set_ylabel(
        'MY OVERALL HEALTH',
        fontproperties=humor_font,
        fontsize=12,
        labelpad=8
    )

    # 添加底部版权文本
    fig.text(
        0.5, 0.05,
        '"Stove Ownership" from xkcd by Randall Munroe',
        ha='center',
        fontproperties=humor_font,
        fontsize=10,
        style='italic'  # 斜体
    )

plt.show()
```

# 2、使用示例：

### 示例 1：风格对比

```python
from matplotlib.font_manager import FontProperties
import os

# --------------------------
# 加载两种目标字体
# --------------------------
# 1. 正常风格字体：SimHei
simhei_path = "C:/Windows/Fonts/simhei.ttf"
simhei_font = FontProperties(fname=simhei_path, size=12)

# 2. xkcd风格字体：FZKaTong-M19S
fzkt_path = "C:/Windows/Fonts/FZKaTong-M19S.ttf"
fzkt_font = FontProperties(fname=fzkt_path, size=12)

# --------------------------
# 生成数据与创建子图
# --------------------------
# 生成带噪声的正弦曲线数据
x = np.linspace(0, 10, 50)
y = np.sin(x) + np.random.randn(50) * 0.2

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5), dpi=100)

# --------------------------
# 子图1：正常风格（SimHei字体）
# --------------------------
# 绘制正常风格折线图
line1 = ax1.plot(x, y, label='随机波动曲线', color='#2E86AB', linewidth=2)
# 设置标题
ax1.set_title('普通风格图表', fontproperties=simhei_font, fontsize=14, fontweight='bold')
# 设置坐标轴标签
ax1.set_xlabel('X轴坐标', fontproperties=simhei_font, fontsize=12)
ax1.set_ylabel('Y轴数值', fontproperties=simhei_font, fontsize=12)
# 设置图例
ax1.legend(prop=simhei_font, fontsize=11, loc='upper right')
# 设置网格
ax1.grid(alpha=0.3, linestyle='--')

# --------------------------
# 子图2：xkcd风格（FZKaTong-M19S字体（方正卡通简体））
# --------------------------
with plt.xkcd(scale=2, length=50, randomness=4):
    # 绘制xkcd风格折线图
    line2 = ax2.plot(x, y, label='随机波动曲线', color='#2E86AB', linewidth=2)
    # 设置标题
    ax2.set_title('xkcd风格图表', fontproperties=fzkt_font, fontsize=14)
    # 设置坐标轴标签
    ax2.set_xlabel('X轴坐标', fontproperties=fzkt_font, fontsize=12)
    ax2.set_ylabel('Y轴数值', fontproperties=fzkt_font, fontsize=12)
    # 设置图例
    ax2.legend(prop=fzkt_font, fontsize=11, loc='upper right')
    # xkcd风格建议不显示网格，避免线条过多影响手绘感（若需显示可保留）
    ax2.grid(alpha=0.3, linestyle='solid')  # 如需网格，用实线避免虚线冲突

plt.tight_layout(pad=3.0)
plt.show()
```

### 示例 2：手绘风格柱状图

```python
# 导入字体管理工具和系统路径模块
from matplotlib.font_manager import FontProperties
import os

# 配置xkcd默认英文手绘字体：Humor Sans
humor_sans_path = "C:/Windows/Fonts/Humor-Sans.ttf"
humor_font = FontProperties(fname=humor_sans_path, size=11)

# 启用xkcd手绘风格（带自定义参数）
with plt.xkcd(scale=1.1,length=100,randomness=2):
    # 全局字体配置
    plt.rcParams['font.family'] = humor_font.get_name()
    plt.rcParams['axes.unicode_minus'] = False
    
    # 图表数据
    months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
    sales = [120, 180, 150, 220, 190, 250]
    
    plt.figure(figsize=(10, 6), dpi=100)
    
    # 绘制xkcd风格柱状图
    bars = plt.bar(
        x=months,
        height=sales,
        color=['#A23B72', '#F18F01', '#C73E1D', '#2E86AB', '#F24236', '#3E92CC'],
        edgecolor='#333333',
        linewidth=1.2
    )
    
    # 为每个柱子添加数值标注
    for bar in bars:
        height = bar.get_height()
        plt.text(
            x=bar.get_x() + bar.get_width()/2,
            y=height + 5,
            s=f'{int(height)}',
            ha='center',
            va='bottom',
            fontsize=10,
        )
    
    
    plt.xlabel('Month', fontsize=12, labelpad=8)
    plt.ylabel('Sales Volume (Units)', fontsize=12, labelpad=8)
    plt.title('2024 H1 Product Sales Trend (Hand-Drawn Style)', fontsize=14, pad=12)
    plt.ylim(0, 300)
    plt.grid(axis='y', alpha=0.3, linestyle='-', linewidth=1)
    plt.tight_layout()
    plt.show()
```
