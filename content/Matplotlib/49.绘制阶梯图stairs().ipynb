```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# 设置中文字体
plt.rcParams['font.sans-serif']=['SimHei']
#显示负号
matplotlib.rcParams['axes.unicode_minus']=False
```

# 1、函数语法：

```python
# 准备数据
values = [25, 28, 26, 30, 27]  # 5个数据
edges = np.arange(1, 7)  # 6个边界，对应5个区间

# 绘制阶梯图
fig, ax = plt.subplots(figsize=(10, 6))
stairs = ax.stairs(
    values,                   # 阶梯的高度值（y轴数据，因orientation设置为horizontal，实际表现为x轴方向）
    edges,                    # 阶梯的边界位置（x轴坐标，因orientation设置为horizontal，实际表现为y轴方向）
    fill=True,                # 启用填充功能，填充阶梯与基线之间的区域
    baseline=20,              # 设置基线位置为20（填充区域从基线延伸到阶梯高度）
    label='数据',             # 设置图例标签，用于在legend中显示
    color='skyblue',          # 设置填充区域的颜色
    edgecolor='navy',         # 设置阶梯线的颜色
    linewidth=3,              # 设置阶梯线的宽度
    alpha=0.5,                # 设置整体透明度（包括填充区域和线条）
    orientation='horizontal'  # 设置阶梯图方向为水平
)

plt.title('基本阶梯图示例')
plt.xlabel('X轴')
plt.ylabel('Y轴')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

# 2、使用示例：

### 示例 1：通过返回对象调整图表样式

```python
# 创建示例数据
np.random.seed(42)
values = np.random.randint(1, 10, 10)
edges = np.arange(len(values) + 1)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6)) 

# 左侧子图：展示初始样式的阶梯图
ax1.stairs(values, edges,label='初始样式')
ax1.set_title('初始阶梯图样式', fontsize=12, fontweight='bold')
ax1.legend()
ax1.grid(alpha=0.3)

# 右侧子图：通过StepPatch对象修改阶梯图样式
step_patch = ax2.stairs(values, edges,label='修改后样式')

# 使用get_data()方法获取StepPatch对象中的原始数据
original_data = step_patch.get_data()
print(f"原始StepPatch的值、边界和基线：{original_data}")
# 使用set_data()方法更新阶梯图数据
step_patch.set_data(values=values-1, edges=edges+1, baseline=2)
new_data = step_patch.get_data()
print(f"更新后StepPatch的值、边界和基线：{new_data}")

# 通过StepPatch对象修改样式
step_patch.set_facecolor('lightblue')  # 设置填充颜色
step_patch.set_edgecolor('red')        # 设置边缘颜色
step_patch.set_linewidth(3)            # 设置线宽
step_patch.set_linestyle('--')         # 设置线型
step_patch.set_alpha(0.5)              # 设置透明度
step_patch.set_hatch('//')             # 设置填充图案
step_patch.set_fill(True)              # 启用填充

ax2.set_title('通过StepPatch修改后样式', fontsize=12, fontweight='bold')
ax2.legend()
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
