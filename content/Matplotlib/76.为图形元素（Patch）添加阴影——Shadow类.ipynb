```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# 设置中文字体
plt.rcParams['font.sans-serif']=['SimHei']
#显示负号
matplotlib.rcParams['axes.unicode_minus']=False
```

# 1、函数语法：

```python
from matplotlib.patches import Rectangle,Shadow

fig, ax = plt.subplots(figsize=(8, 6))

# 创建一个矩形补丁
rect = Rectangle((0.2, 0.2), 0.6, 0.5, facecolor='blue', alpha=0.7)

# 创建阴影效果
shadow = Shadow(patch=rect, # 指定要添加阴影的图形元素
                ox=0.03,    # 阴影在x轴方向的偏移量
                oy=-0.03,   # 阴影在y轴方向的偏移量
                shade=0.4   # 阴影的暗度（0-1之间，值越小阴影越浅）
               )

# 将阴影和矩形添加到图表中
ax.add_patch(shadow)
ax.add_patch(rect)

ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
plt.title('基本阴影效果示例')
plt.show()
```

# 2、使用示例：

### 示例 1：偏移量设置

```python
from matplotlib.patches import Rectangle,Shadow

# 不同方向的阴影效果
shadows = {
    '右下': (0.02, -0.02),
    '左上': (-0.02, 0.02),
    '正下': (0, -0.03)
}

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# 遍历阴影参数和子图对象，逐个绘制带阴影的矩形
for (name, (ox, oy)), ax in zip(shadows.items(), axes):
    rect = Rectangle((0.2, 0.2), 0.6, 0.5, facecolor='blue', alpha=0.7)
    shadow = Shadow(rect, ox, oy, shade=0.5)
    
    ax.add_patch(shadow)
    ax.add_patch(rect)
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_aspect('equal')
    ax.set_title(f'{name}阴影')

plt.tight_layout()
plt.show()
```

### 示例 2：自定义阴影

```python
from matplotlib.patches import Rectangle,Shadow

fig, ax = plt.subplots(figsize=(6, 6))

# 创建按钮状图表元素
buttons = [
    {'pos': (0.3, 0.7), 'label': '按钮 1', 'color': 'blue'},
    {'pos': (0.3, 0.5), 'label': '按钮 2', 'color': 'green'},
    {'pos': (0.3, 0.3), 'label': '按钮 3', 'color': 'coral'}
]

# 循环遍历每个按钮配置，绘制带阴影的按钮
for btn in buttons:
    x, y = btn['pos']
    rect = Rectangle((x, y), 0.4, 0.15, facecolor='lightblue')
    shadow = Shadow(rect, 
                    0.02, 
                    -0.02,
                    shade=0.7, 
                    fc=btn['color'],    # 自定义阴影填充色 
                    edgecolor='black',  # 阴影边框颜色
                    linewidth=4,        # 阴影边框宽度
                    alpha=0.5           # 阴影透明度
                   )
    
    ax.add_patch(shadow)
    ax.add_patch(rect)
    ax.text(x+0.2, y+0.075, btn['label'], ha='center', va='center', fontsize=12)

ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_title('阴影颜色和透明度')
ax.set_aspect('equal')
ax.axis('off')
plt.show()
```
