### 在使用 matplotlib 时，我们使用以下导入约定：

```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
%matplotlib inline
```

```python
# linspace(start,end,step)函数可以创建指定区间均匀间隔的数字数组
x = np.linspace(0 , 4*np.pi , 500)
y = np.sin(x)

plt.plot(x,y)
```

##### 隐藏 <matplotlib.lines.Line2D>输出，有以下两种方式

```python
x = np.linspace(0 , 4*np.pi , 500)
y = np.sin(x)

plt.plot(x,y)
# 一种方法是在绘图代码的末尾使用plt.show()函数。
# 如果要生成许多子图或在单个子图上绘制许多图，则键入的符号会更少
plt.show()
```

```python
x = np.linspace(0 , 4*np.pi , 500)
y = np.sin(x)

# 使用分号; 结束该行。这会在生成绘图时抑制不需要的输出
# 一般来说，使用分号会阻止 IPython 从代码块的该行打印任何输出值
plt.plot(x,y);
```

### 在绘制图形时，可以在一组图形内绘制多组数据

```python
x = np.linspace(0 , 4*np.pi , 500)
y = np.sin(x)
y_1 = np.cos(x)

# 使用两个plot()函数绘制两组数据
plt.plot(x,y)
plt.plot(x,y_1)
plt.show()
```

```python
x = np.linspace(0 , 4*np.pi , 500)
y = np.sin(x)
y_1 = np.cos(x)

# 可以使用一个plot()函数
plt.plot(x,y , x,y_1)
plt.show()
```
