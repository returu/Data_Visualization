```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# 设置中文字体
plt.rcParams['font.sans-serif']=['SimHei']
#显示负号
matplotlib.rcParams['axes.unicode_minus']=False
```

# 1、函数语法：

```python
import matplotlib.patches as patches

# 创建图形和坐标轴
fig, ax = plt.subplots(figsize=(8, 6))

# 创建一个矩形  
rect = patches.Rectangle((0.2, 0.4),           # 左下角坐标
                        0.5,                   # 宽度
                        0.3,                   # 高度
                        angle=30,              # 旋转角度
                        linewidth=2,           # 边框粗细
                        linestyle='--',        # 虚线边框
                        edgecolor='r',         # 边框颜色
                        facecolor='lightblue', # 填充颜色
                        alpha=0.6)             # 透明度

# 将矩形添加到坐标轴
ax.add_patch(rect)

# 设置坐标轴范围
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

plt.title('基本矩形示例')
plt.show()
```

# 2、使用示例：

### 示例 1：突出显示数据区域

```python
from matplotlib.patches import Rectangle

# 生成模拟数据（带异常波动的折线图）
x = np.arange(0, 20, 1)  # x轴数据：0-19
y = np.sin(x) + np.random.normal(0, 0.2, size=20) 
abnormal_x = [9, 12]  # 异常数据区间

fig, ax = plt.subplots(figsize=(12, 6))

# 绘制折线图
ax.plot(x, y, color='#2E86AB', linewidth=2, marker='o', markersize=4, label='Data Trend')

# 用矩形标注异常区间
rect = Rectangle(
    xy=(abnormal_x[0], -2),
    width=abnormal_x[1] - abnormal_x[0],
    height=4.0, 
    facecolor='#F24236',
    alpha=0.2,
    label='异常区间'
)
ax.add_patch(rect)

# 添加标注文本
ax.text(
    x=(abnormal_x[0] + abnormal_x[1])/2,
    y=1,
    s='异常数据区间',
    ha='center', 
    va='center',
    fontsize=14,
    color='#D62728',
    fontweight='bold'
)


ax.set_xlim(0, 19)
ax.set_ylim(-2, 2)
ax.set_xlabel('Time (s)', fontsize=12)
ax.set_ylabel('Value', fontsize=12)
ax.set_title('突出显示数据区域', fontsize=14, pad=10)
ax.legend(fontsize=10)
ax.grid(True, alpha=0.3)

plt.show()
```

### 示例 2：在图像内绘制矩形

```python
from matplotlib.patches import Rectangle
import matplotlib.image as img

# 读取图像
image_path = "example.jpg"
image = img.imread(image_path)

fig, ax = plt.subplots(figsize=(8, 6))
ax.imshow(image)

# 定义矩形基础参数
rect_xy = (350, 120)    # 矩形左上角坐标
rect_width = 150        # 水平宽度
rect_height = 300       # 垂直高度
rotate_angle = 40       # 旋转角度

# 计算旋转中心（此处用矩形中心，对应rotation_point="center"）
center_x = rect_xy[0] + rect_width / 2
center_y = rect_xy[1] + rect_height / 2

# 创建旋转矩形
rotated_rect = Rectangle(
    xy=rect_xy,
    width=rect_width,
    height=rect_height,
    angle=rotate_angle, 
    rotation_point="center", 
    edgecolor="red",
    linewidth=3,
    linestyle="-",
    facecolor="none"
)
ax.add_patch(rotated_rect)

# 标注旋转中心
ax.scatter(center_x, center_y, color="red", s=60, zorder=5, label="Rotation Center")
ax.legend(fontsize=10)


ax.axis("off")
ax.set_title(f"Rotated Rectangle (Angle: {rotate_angle}°) on Image", fontsize=14, pad=10)
plt.tight_layout()
plt.show()
```
