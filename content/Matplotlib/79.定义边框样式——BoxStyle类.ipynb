# 2、使用示例：

### 示例 1：在FancyBboxPatch中使用

```python
from matplotlib.patches import FancyBboxPatch,BoxStyle

# 准备画布
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
ax.set_xlim(0, 10)
ax.set_ylim(0, 8)
ax.set_aspect('equal')

# 定义一些样式和位置
styles = [
    ('Square', BoxStyle.Square(pad=0.3), 1, 7),
    ('Circle',BoxStyle.Circle(pad=0.3),4, 7),
    ('Ellipse',BoxStyle.Ellipse(pad=0.3), 7, 7),
    ('LArrow',BoxStyle.LArrow(pad=0.3),1, 5),
    ('RArrow',BoxStyle.RArrow(pad=0.3), 4, 5),
    ('DArrow',BoxStyle.DArrow(pad=0.1), 7, 5),
    ('Round',BoxStyle.Round(pad=0.3,rounding_size=0.5),1, 3),
    ('Round4',BoxStyle.Round4(pad=0.3,rounding_size=0.5),4, 3),
    ('Sawtooth',BoxStyle.Sawtooth(pad=0.3,tooth_size=0.2),1, 1),
    ('Roundtooth',BoxStyle.Roundtooth(pad=0.3,tooth_size=0.2),4, 1)
]

# 绘制各种边框样式
for i, (name, style, x, y) in enumerate(styles):
    # 创建FancyBboxPatch对象
    fancy_bbox = FancyBboxPatch((x-0.4, y-0.2), 0.8, 0.4,
                                boxstyle=style,
                                facecolor='lightblue',
                                edgecolor='navy',
                                linewidth=1.5,
                                alpha=0.7)
    ax.add_patch(fancy_bbox)
    ax.text(x, y, name, 
            ha='center', va='center', fontsize=12)

ax.set_title('FancyBboxPatch - BoxStyle 示例')
ax.axis('off')
plt.show()
```

### 示例 2：在文本注释中使用

```python
from matplotlib.patches import BoxStyle

fig, ax = plt.subplots(figsize=(10, 6))

# 示例数据
x = [1, 2, 3, 4, 5]
y = [2, 5, 3, 7, 4]

# 美化折线图
line = ax.plot(x, y, 'o-', alpha=0.8, linewidth=2.5, markersize=8, 
               color='#2E86AB', markerfacecolor='#A23B72', 
               markeredgecolor='white', markeredgewidth=1.5)

# 设置坐标轴样式
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_alpha(0.3)
ax.spines['bottom'].set_alpha(0.3)

# 设置网格线
ax.grid(True, alpha=0.3, linestyle='--', linewidth=0.5)

# 美化文本框标注
ax.text(2, 5.7, "峰值点", size=13, va="bottom", ha="center", fontweight='bold',
        bbox=dict(boxstyle=BoxStyle("Round4,pad=0.5"), 
                 fc="#A8E6CF", ec="#1B5E20", alpha=0.9, 
                 linewidth=1.5))

ax.text(4.4, 7.5, "最高点", size=13, va="bottom", ha="center", fontweight='bold',
        bbox=dict(boxstyle=BoxStyle("RArrow,pad=0.4"), 
                 fc="#FFAAA5", ec="#D32F2F", alpha=0.9, 
                 linewidth=1.5))

ax.text(3.2, 2.5, "回调点", size=11, va="top", ha="left", fontweight='bold',
        bbox=dict(boxstyle=BoxStyle("Circle,pad=0.4"), 
                 fc="#FFD3B6", ec="#FF8B00", alpha=0.9, 
                 linewidth=1.5))

# 添加一个额外的样式示例
ax.text(1, 1.5, "起点", size=11, va="center", ha="right", fontweight='bold',
        bbox=dict(boxstyle=BoxStyle("Sawtooth,pad=0.4,tooth_size=0.1"), 
                 fc="#D4A5A5", ec="#6D6875", alpha=0.9, 
                 linewidth=1.5))

# 添加箭头连接文本和对应的数据点，增强关联性
ax.annotate('', xy=(2, 5), xytext=(2, 5.5),
            arrowprops=dict(facecolor='#27ae60', shrink=0.05, width=1.5, headwidth=5))

ax.annotate('', xy=(4, 7), xytext=(4.2, 7.5),
            arrowprops=dict(facecolor='#c0392b', shrink=0.05, width=1.5, headwidth=5))

ax.annotate('', xy=(3, 3), xytext=(3.2, 2.5),
            arrowprops=dict(facecolor='#f39c12', shrink=0.05, width=1.5, headwidth=5))


# 设置坐标轴范围
ax.set_ylim(0, 8.5)
ax.set_xlim(0.5, 5.5)

# 添加标题和标签
ax.set_title('BoxStyle 文本框样式应用示例', fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel('X轴', fontsize=12, labelpad=10)
ax.set_ylabel('Y轴', fontsize=12, labelpad=10)

# 调整刻度样式
ax.tick_params(axis='both', which='major', labelsize=10)

# 添加背景色
ax.set_facecolor('#F8F9FA')
fig.patch.set_facecolor('white')

# 添加图例
ax.legend(['数据趋势线'], loc='upper left', frameon=True, 
          facecolor='white', edgecolor='gray', framealpha=0.8)

# 调整布局
plt.tight_layout()

plt.show()
```
