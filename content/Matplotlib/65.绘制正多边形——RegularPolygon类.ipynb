```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# 设置中文字体
plt.rcParams['font.sans-serif']=['SimHei']
#显示负号
matplotlib.rcParams['axes.unicode_minus']=False
```

# 1、函数语法：

```python
from matplotlib.patches import RegularPolygon

fig, ax = plt.subplots()
ax.set_aspect('equal')

# 绘制正三角形
triangle = RegularPolygon((2, 2),                # 中心坐标
                          3,                     # 边数
                          radius=1,              # 半径
                          orientation=np.pi/6,   # 旋转角度
                          facecolor='red',       # 填充颜色
                          alpha=0.6,             # 透明度
                          edgecolor='black'      # 边框颜色
                         )

ax.add_patch(triangle)
ax.set_xlim(0, 4)
ax.set_ylim(0, 4)
plt.title('正三角形')
plt.show()
```

# 2、使用示例：

### 示例 1：六边形网格地图（蜂巢图）

```python
from matplotlib.patches import RegularPolygon

fig, ax = plt.subplots()
# 设置坐标轴比例相等，确保六边形不会因拉伸变形
ax.set_aspect('equal')

radius = 1.0
# 双重循环生成网格：i控制x方向，j控制y方向
for i in range(-5, 6):
    for j in range(-5, 6):
        # 六边形网格坐标计算
        x = i * 1.5 * radius
        y = (j + 0.5 * (i % 2)) * np.sqrt(3) * radius
        
        # 创建六边形对象
        hexagon = RegularPolygon(
            xy=(x, y),                                   # 六边形中心坐标
            numVertices=6,                               # 6条边
            radius=radius,                               # 半径
            orientation=np.pi/6,                         # 旋转30°（π/6弧度），让上下边水平
            facecolor=plt.cm.Blues((i + j + 10) / 20),   # 颜色映射，动态变化
            edgecolor='black',                           # 边框
            alpha=0.8                                    # 透明度
        )
        ax.add_patch(hexagon)

plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.title('六边形网格地图')
plt.axis('off')
plt.show()
```

### 示例 2：艺术创作与装饰元素

```python
from matplotlib.patches import RegularPolygon

fig, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect('equal')
# 设置整个画布的背景色
fig.patch.set_facecolor('#1a1a2e')

# 创建螺旋状多边形图案
n = 20
for i in range(n):
    # 计算当前多边形的角度，实现围绕中心旋转的效果，i越大旋转角度越大
    angle = i * 2 * np.pi / n
    # 计算距离中心的距离，现从中心向外扩散的螺旋轨迹
    distance = 0.1 + 0.9 * i / n
    # 计算多边形大小，实现"外小内大"的视觉层次感
    size = 0.5 * (1 - i/n)
    
    # 创建正五边形对象
    poly = RegularPolygon(
        xy=(distance * np.cos(angle), distance * np.sin(angle)),
        numVertices=5,
        radius=size,
        orientation=angle,
        facecolor=plt.cm.hsv(i/n),
        edgecolor='white',
        alpha=0.8
    )
    ax.add_patch(poly)

plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
plt.title('螺旋多边形艺术图案', color='white')
plt.axis('off')
plt.show()
```
